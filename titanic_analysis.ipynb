```python
"""
Titanic Data Analysis and JSON Export
Author: David Gill
Description: Analyze Titanic passenger data, engineer features, and export to JSON
"""
 
import pandas as pd
import numpy as np
import json
from pathlib import Path
 
# Set up paths
DATA_DIR = Path("data")
CSV_FILE = DATA_DIR / "titanic.csv"
JSON_FILE = DATA_DIR / "titanic_data.json"
 
# Create data directory if it doesn't exist
DATA_DIR.mkdir(exist_ok=True)
 
print("Project setup complete!")
print(f"Data directory: {DATA_DIR}")
print(f"CSV file location: {CSV_FILE}")
```

```python
# Load dataset
df = pd.read_csv(CSV_FILE)

# Preview data
print(df.head())
```

```python
# Basic info about dataset
print(df.info())

# Summary statistics
print(df.describe())
```

```python
# Drop Cabin column
df = df.drop(columns=["Cabin"])

print(df.columns)
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
# Fill missing Embarked values with the most common value (mode)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

print(df["Embarked"].isnull().sum())
```
